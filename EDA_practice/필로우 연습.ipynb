```python
from PIL import Image, ImageColor
import os
```

# 이미지 열기

```python
img_path = os.getenv("HOME") + "/Desktop/image/kwon.jpeg"
img = Image.open(img_path)                                       # 이미지 열기
print(type(img))
img
```

```python
w, h = img.size
print((w,h))
```

# 이미지 포맷, 사이즈, 채널 확인

```python
print(img.format)              # 포맷, 사이즈, 모드 확인
print(img.size)
print(img.mode)
```

# 이미지 자르기

```python
img.crop((300,250,800,700))     #left-top 가로 세로 시작점,  right-bottom 가로 세로 끝점
```

# 이미지 저장

```python
cropped_img_path = os.getenv("HOME") + "/Desktop/image/cropped_img.jpg"
img.crop((300,250,800,700)).save(cropped_img_path)                             # 이미지 저장
```

# 이미지 행렬로 변환

```python
import numpy as np
img_arr = np.array(img)      # 그냥 이미지 파일을 np.array() 에 넣어버리면 끝
print(img_arr.shape)
print(img_arr.ndim)
img_arr
```
