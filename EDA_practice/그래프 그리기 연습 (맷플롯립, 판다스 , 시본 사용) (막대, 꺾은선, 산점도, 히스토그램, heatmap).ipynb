# 선 그래프 그리기

```python
from datetime import datetime
import pandas as pd
import os
```

```python
csv_path = os.getenv("HOME") + "/Desktop/data/AMZN.csv"
data = pd.read_csv(csv_path, index_col=0, parse_dates=True)
data
```

```python
price = data['Close']
```

## 축 그리기, 좌표축 설정

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
price.plot(ax=ax, style='black')
plt.ylim([1600,2200])
plt.xlim(['2019-05-01', '2020-03-01'])
```

## 주석 달기   
   
고급 기법

```python
important_data = [(datetime(2019, 6, 3), "Low Price"), (datetime(2020, 2, 19), "Peak Price")]
for d, label in important_data:
    ax.annotate(label, xy = (d, price.asof(d)+10),
                xytext = (d, price.asof(d) +100),
                arrowprops = dict(facecolor='red'))
```

## 그리드, 타이틀 달기

```python
plt.grid()
ax.set_title("StockPrice")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
price.plot(ax=ax, style='black')         # 판다스의 선 그래프 그리기 기능 plot 사용하면서 matplotlib에서 정의한 subplot 공간 ax 사용
plt.ylim([1600,2200])
plt.xlim(['2019-05-01', '2020-03-01'])

important_data = [(datetime(2019, 6, 3), "Low Price"), (datetime(2020, 2, 19), "Peak Price")]
for d, label in important_data:
    ax.annotate(label, xy = (d, price.asof(d)+10),
                xytext = (d, price.asof(d) +100),
                arrowprops = dict(facecolor='red'))

plt.grid()
ax.set_title("StockPrice")

plt.show()
```

# 선그래프 라인 스타일

```python
x = np.linspace(0, 10, 100) 

plt.plot(x, x + 0, linestyle='solid') 
plt.plot(x, x + 1, linestyle='dashed') 
plt.plot(x, x + 2, linestyle='dashdot') 
plt.plot(x, x + 3, linestyle='dotted')
plt.plot(x, x + 0, '-g') # solid green 
plt.plot(x, x + 1, '--c') # dashed cyan 
plt.plot(x, x + 2, '-.k') # dashdot black 
plt.plot(x, x + 3, ':r'); # dotted red
plt.plot(x, x + 4, linestyle='-') # solid 
plt.plot(x, x + 5, linestyle='--') # dashed 
plt.plot(x, x + 6, linestyle='-.') # dashdot 
plt.plot(x, x + 7, linestyle=':'); # dotted
```


### 2. seaborn 과 matplotlib 사용

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")     # seaborn의 load_dataset 메서드 사용
df = pd.DataFrame(tips)

sns.barplot(data=df, x='sex', y = 'tip')            # seaborn의 .barplot() 메서드 사용   데이터프레임만 있으면 한줄로 끝.
```


**Matplot과 함께 사용하여 figsize, title을 정하는 등 그래프에 다양한 옵션 사용 가능**

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='sex', y='tip')
plt.ylim(0, 4)
plt.title('Tip by sex')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='day', y='tip')                    # 요일에 따른 팁 그래프
plt.ylim(0, 4)
plt.title('Tip by day')
```

```python
fig = plt.figure(figsize=(13,10))
ax1 = fig.add_subplot(2,2,1)                                     # subplot 도 사용 가능.
sns.barplot(data=df, x='day', y='tip',palette="ch:.25")
ax2 = fig.add_subplot(2,2,2)
sns.barplot(data=df, x='sex', y='tip')
```
