### 전설 이름 토큰화

```python
import re
def tokenize(name):
    name_split = name.split(" ")
    
    tokens = []
    for part_name in name_split:
        a = re.findall('[A-Z][a-z]*', part_name)
        tokens.extend(a)
        
    return np.array(tokens)
```

```python
name = "CharizardMega Charizard X"
tokenize(name)
```

```python
all_tokens = list(legendary["Name"].apply(tokenize).values)

token_set = []
for token in all_tokens:
    token_set.extend(token)

print(len(set(token_set)))
print(token_set)
```

### 많이 사용된 토큰을 추출 (collections 모듈의 Counter 사용)

```python
from collections import Counter
most_common = Counter(token_set).most_common(10)
most_common
```

이제 마지막으로 이렇게 전설의 포켓몬 이름에 등장하는 토큰이 포켓몬의 이름에 있는지의 여부를 나타내는 컬럼을 만들겠습니다.

```python
for token, _ in most_common:
    pokemon[token] = pokemon["Name"].str.contains(token)

pokemon.head(10)
```

```python
pokemon.columns
```

이제 전설의 포켓몬이 많이 가지는 Forme와 같은 토큰의 컬럼 값이 True라면 그 포켓몬은 전설의 포켓몬일 확률이 높다고 판단할 수 있겠군요!

## feature engineering, 전처리 안 한 데이터로 일단 돌려보기

```python
print(original_data.shape)
original_data.head()
```

```python
features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']
```

```python
target = 'Legendary'
```

```python
X = original_data[features]
print(X.shape)
X.head()
```

```python
y = original_data[target]
print(y.shape)
y.head()
```

**훈련데이터 테스트데이터 분리**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=25)   # 여기에 random 이 왜 필요하지?
model
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
